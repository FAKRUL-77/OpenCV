```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("images/fakrul.png", 1)

cv.imshow('fakrul', img)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

**To hide xticks, and yticks**   
**#-->> plt.xticks([]), plt.yticks([])**

```python
plt.imshow(img)
plt.xticks([]), plt.yticks([])
plt.show()
```
