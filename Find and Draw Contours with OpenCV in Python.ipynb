```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("images/baseball1.png")
img = cv.resize(img, (512, 512))
g_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(g_img, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
print("Number of contours = " + str(len(contours)))

cv.drawContours(img, contours, -1, (0, 255, 0), 3)

plt.imshow(img)
cv.imshow("image", img)
cv.imshow("image_gray", g_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

**# cv.findContours(threshold, Contour mode, Contour approximation method)**  
**contours :** is a python list of all the contours in the image. Each individual contour is a numpy array of (X, Y) coordinates of the boundary points of the object.  
**hierarchy :** is the optional vector containing the information about image topology.  
**drawContours(img, contours, -1, (0, 255, 0), 3)**
