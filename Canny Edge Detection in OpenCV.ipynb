**The Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images. It was developed by John F. Canny in 1986. Canny also produced a computational theory of edge detection explaining why the technique works.**

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
rules = cv.imread("images/canny_edge_detection.PNG")
plt.imshow(rules)
plt.show()
```

```python
img = cv.imread('images/fakrul.png', 0)

canny = cv.Canny(img, 100, 200)

titles = ['image', 'canny']
images = [img, canny]

cv.imshow('image', img)
cv.imshow('canny', canny)


cv.waitKey(0)
cv.destroyAllWindows()

for i in range(2):
    plt.subplot(1, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

```python
img = cv.imread('images/sudoko.jpg', cv.IMREAD_GRAYSCALE)

canny = cv.Canny(img, 50, 200)
lap = cv.Laplacian(img, cv.CV_64F, ksize=1)
lap = np.uint8(np.absolute(lap))

sobelX = cv.Sobel(img, cv.CV_64F, 1, 0)
sobelY = cv.Sobel(img, cv.CV_64F, 0, 1)

sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))

sobelCombined = cv.bitwise_or(sobelX, sobelY)


titles = ['image', 'lap_gradient', 'sobelX', 'sobelY', 'sobelCombined', 'canny']
images = [img, lap, sobelX, sobelY, sobelCombined, canny]

cv.imshow('image', img)
cv.imshow('lap_gradient', lap)
cv.imshow('sobelX', sobelX)
cv.imshow('sobelY', sobelY)
cv.imshow('sobelCombined', sobelCombined)
cv.imshow('canny', canny)


cv.waitKey(0)
cv.destroyAllWindows()

for i in range(6):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
