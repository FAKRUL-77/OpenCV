```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread("images/fakrul.png")

kernel = np.ones((5, 5), np.float32)/25
dst = cv.filter2D(img, -1, kernel)
blur = cv.blur(img, (5, 5))
gblur = cv.GaussianBlur(img, (5 ,5), 0)
mblur = cv.medianBlur(img, 3)
bilateralFilter = cv.bilateralFilter(img, 9, 75, 75)

cv.imshow('image', img)
cv.imshow('filter_image', dst)
cv.imshow('blur_image', blur)
cv.imshow('gblur_image', gblur)
cv.imshow('mblur_image', mblur)
cv.imshow('biF_image', bilateralFilter)

cv.waitKey(0)
cv.destroyAllWindows()

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
dst = cv.cvtColor(dst, cv.COLOR_BGR2RGB)
blur = cv.cvtColor(blur, cv.COLOR_BGR2RGB)
gblur = cv.cvtColor(gblur, cv.COLOR_BGR2RGB)
mblur = cv.cvtColor(mblur, cv.COLOR_BGR2RGB)
bilateralFilter = cv.cvtColor(bilateralFilter, cv.COLOR_BGR2RGB)


titles = ['Fakrul', '2D filter', 'blur', 'gblur', 'mblur', 'biFilter']
images = [img, dst, blur, gblur, mblur, bilateralFilter]

for i in range(6):
    
    plt.subplot(2, 3, i+1), plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```

**filter2D :** use for homogeneous filter and argements are (source, desire_depth, kernel)  
**GaussianBlur :** cv2.GaussianBlur(source, kernal, sigma x value)  
**bilateralFilter :** cv.bilateralFilter(source, diameter_of_each_pixel_neighborhood, sigma_color, sigma_space)
