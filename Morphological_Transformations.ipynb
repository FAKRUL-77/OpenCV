```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('images/chocolet2.jpg', cv.IMREAD_GRAYSCALE)
_, mask = cv.threshold(img, 220, 255, cv.THRESH_BINARY_INV)

kernal = np.ones((5,5), np.uint8)
dilation = cv.dilate(mask, kernal, iterations=3)
erosion = cv.erode(mask, kernal, iterations=1)
opening = cv.morphologyEx(mask, cv.MORPH_OPEN, kernal)
closing = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernal)
morphological_gradient = cv.morphologyEx(mask, cv.MORPH_GRADIENT, kernal)
top_hat = cv.morphologyEx(mask, cv.MORPH_TOPHAT, kernal)


titles = ['image', 'mask', 'dilation', 'erosion', 'opening', 'closing', 'm_g', 't_h']
images = [img, mask, dilation, erosion, opening, closing, morphological_gradient, top_hat]

cv.imshow("mask", mask)
cv.imshow("image", img)
cv.imshow("dilation", dilation)
cv.imshow("erosion", erosion)
cv.imshow("opening", opening)
cv.imshow("closing", closing)
cv.imshow("m_g", morphological_gradient)
cv.imshow("t_h", top_hat)



cv.waitKey(0)
cv.destroyAllWindows()

for i in range(8):
    ax = plt.subplot(3, 2, i+1), plt.imshow(images[i], "gray")
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```
